```python
def get_numbers():
    """ Get numbers from the user.

    Returns:
        String: user's numbers.
    """
    return input('‚ö† Please, enter the numbers: ').split()
```

```python
def check_numbers(numbers):
    """ Perform and show the operations of maximum, minimum and average on the list of numbers.

    Args:
        numbers(int): List of numbers.
    """
    high, low, avr = numbers[0], numbers[0], 0
    for i in numbers:
        if i > high: high = i
        if i < low: low = i
        avr += i
    avr /= len(numbers) 
    print(''.join((f'The maximun value of {numbers} is: {high} \n',
                   f'The minimun value of {numbers} is: {low} \n',
                   f'The average value of {numbers} is: {avr}')))
```

```python
def get_1_out_factorial(num):
    """ Get 1 / factorial of the number.

    Args:
        num(int): The number to apply the factorial.

    Returns:
        int: 1 divided by the factorial.
    """
    return num if num == 1 else (1 / num * get_1_out_factorial(num - 1))
```

```python
def show_napier(num):
    """ Show Napier's constant.

    Args:
        num(int): Number of terms for Napier's constant.
    """
    napier = 0
    for num in range(1, num + 1):
        napier += get_1_out_factorial(num)  
    print(f'e({num}) is:','{0:.5f}'.format(napier + 1))
```

```python
num = 20    
show_napier(num)
```

‚úÖ Nested functions.

```python
lst = []

def func1(a):
    """ Call the other functions.

    Args:
        a(List): Numbers list.

    Returns:
        Generator object: Grouped list of values divisible by 5 and 7.
    """    
    def func2():
        """ Stores numbers divisible by 5 in a list.

        Args:
            None.

        Returns:
            Generator object: Numbers divisible by 5.
        """
        for i in a:
            if i % 5 == 0:
                yield i   
            
    def func3():
        """ Stores numbers divisible by 7 in a list.

        Args:
            None.

        Returns:
            Generator object: Numbers divisible by 7.
        """
        for i in a:
            if i % 7 == 0:
                yield i
                
    yield sorted(list(func2()) + list(func3()))
```

```python
# Add numbers from 1 to 100 to the list.
[lst.append(i) for i in range(1, 101)]

print(f'List: {lst} \n')    
print(f'Result: {list(func1(lst))}')
```

‚úÖ Send values to the student class constructor.

```python
class Student:
    """ Student Class. 
    """
#%%    
    def __init__(self, stu_id, name, eng_quiz, math_quiz, science_quiz):
        """
        Default Constructor.

        Parameters
        ----------
        stu_id : String
            Student's ID.
        name : int
            Student's name.
        eng_quiz : List
            Student's english quiz score.
        math_quiz : List
            Student's math quiz score.
        science_quiz : List
            Student's science quiz score.

        Returns
        -------
        None.

        """
        self._stu_id       = int(stu_id)
        self._name         = name
        self._eng_quiz     = [eng_quiz]
        self._math_quiz    = [math_quiz]
        self._science_quiz = [science_quiz]

#%%
    @property
    def stu_id(self):
        """
        Getter for student's ID.

        Parameters
        ----------
        None.

        Returns
        -------
        int: Student's ID.

        """
        return self._stu_id
    
    @stu_id.setter
    def stu_id(self, stu_id):
        """
        Setter for student's ID.

        Parameters
        ----------
        stu_id : int
            Student's ID.

        Returns
        -------
        int: The student's ID.

        """
        self._stu_id = stu_id

#%%
    @property
    def name(self):
        """
        Getter for student's name.

        Parameters
        ----------
        None.

        Returns
        -------
        String: Student's name.

        """
        return self._name

    @name.setter
    def name(self, name):
        """
        Setter for student's name.

        Parameters
        ----------
        name : String
            Student's name.

        Returns
        -------
        String: The student's name.

        """
        self._name = name

#%%
    @property
    def eng_quiz(self):
        """
        Getter for student's english quiz score.

        Parameters
        ----------
        None.

        Returns
        -------
        List: Student's english quiz score.

        """
        return self._eng_quiz
    
    @eng_quiz.setter
    def eng_quiz(self, eng_quiz):
        """
        Setter for student's english quiz score.

        Parameters
        ----------
        eng_quiz : List
            Student's english quiz score.

        Returns
        -------
        List: The student's english quiz score.

        """
        self._eng_quiz = eng_quiz

#%%
    @property
    def math_quiz(self):
        """
        Getter for student's math quiz score.

        Parameters
        ----------
        None.

        Returns
        -------
        List: Student's math quiz score.

        """
        return self._math_quiz

    @math_quiz.setter
    def math_quiz(self, math_quiz):
        """
        Setter for student's math quiz score.

        Parameters
        ----------
        math_quiz : List
            Student's english math score.

        Returns
        -------
        List: The student's math quiz score.

        """
        self._math_quiz = math_quiz

#%%
    @property
    def science_quiz(self):
        """
        Getter for student's science quiz score.

        Parameters
        ----------
        None.

        Returns
        -------
        List: Student's science quiz score.

        """
        return self._science_quiz
    
    @science_quiz.setter
    def science_quiz(self, science_quiz):
        """
        Setter for student's science quiz score.

        Parameters
        ----------
        science_quiz : List
            Student's science score.

        Returns
        -------
        List: The student's science quiz score.

        """
        self._science_quiz = science_quiz

#%%
    def get_total_score(self):
        """
        Get the sum of the grades of the three subjects of the student.

        Parameters
        ----------
        None.

        Returns
        -------
        int: The sum of the grades of the three subjects of the student.

        """
        return sum(self.eng_quiz + self.math_quiz + self.science_quiz)

#%%
    def get_avg_score(self):
        """
        Get the average of the grades of the three subjects of the student.

        Parameters
        ----------
        None.

        Returns
        -------
        float: The average of the grades of the three subjects of the student.

        """
        return self.get_total_score() / 3

#%%
    def __str__(self):
        """
        Students information output.

        Parameters
        ----------
        None.

        Returns
        -------
        None.

        """
        print(''.join((f'Name: {self._name}\n',
                       f'ID: {self._stu_id}\n',
                       f'English quiz: {sum(self.eng_quiz)}\n',
                       f'Mathematics quiz: {sum(self.math_quiz)}\n',
                       f'Science quiz: {sum(self.science_quiz)}\n')))
    
my_stu = Student(20213093, 'David Doe', 90, 95, 100)
my_stu.__str__()
print(f'Total of the sum of the three subjects: {my_stu.get_total_score()}')
print(f'Average of the three subjects: {my_stu.get_avg_score()}')
```

```python
num = 5

def addition(num):
    """ Add from 1 to num.

    Args:
        num(int): Maximum number of the sum.

    Returns:
        int: The sum of 1 up to num.
    """
    return num if num == 1 else (num + addition(num - 1))
    
print(f'The sum from 1 to {num} is: {addition(num)}')  
```

<font color='blue'>Q2.</font> Funci√≥n recursiva para realizar la potencia de 2 ^ n

Python tiene el operador ** el cual indica una potencia, sin embargo, tomemos a x y n como entradas sin usar ese operador y use una funcion recursiva que retorne x a la potencia de n. Luego, calcule 2 ^ 10 poniendo x como 2 y n como 10.

```python
base     = 2
exponent = 10

def power(base, exponent):
    """ Find the power of a base and exponent.

    Args:
        base(int)    : Power base number.
        exponent(int): Power exponent number.

    Returns:
        int: The result of the power.
    """
    return base if exponent == 1 else (base * power(base, exponent - 1))
    
print(f'The {base} to the power of {exponent} is: {power(base, exponent)}')  
```
