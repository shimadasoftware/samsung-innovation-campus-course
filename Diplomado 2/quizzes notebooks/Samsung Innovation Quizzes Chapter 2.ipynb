<font color='blue'>Q1.</font> Función para hallar la cadena más corta de una lista

```python
#s_list  = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
s_list  = ['bcdefg', 'abc', 'bcd', 'abba', 'cddc', 'opq']

def find_minimum(s_list, minimum):
    """ Find the shortest string in a list.

        Args:
            s_list (List) : List with strings.
            minimum (List): The first string in the list.

        Returns:
            string: The shortest string.
    """
    for e in s_list:
        if len(e) < len(minimum):
            minimum = e        
    return minimum

print(''.join((f'The shortest string in the list {s_list} is:\n',
               f'{find_minimum(s_list, s_list[0])}')), '\n')
```

<font color='blue'>Q2.</font> Función para hallar la cadena más larga de una lista

```python
s_list  = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']

def find_maximum(s_list, maximum):
    """ Find the longest string in a list.

    Args:
        s_list (List) : List with strings.
        maximum (List): The first string in the list.

    Returns:
        string: The longest string.
    """
    for e in s_list:
        if len(e) > len(maximum):
            maximum = e        
    return maximum

print(''.join((f'The longest string in the list {s_list} is:\n',
               f'{find_maximum(s_list, s_list[0])}')), '\n')
```

```python
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
result = {} 

s_list.sort(reverse=False, key=len)
print(f'The list: {s_list} \n')

def list_lenght(s_list):
    """ Saves strings as keys and string lengths as values in a dictionary.

    Args:
        s_list (List): List with strings.
    """
    for e in s_list:
        if len(e) in result.keys():
            result[len(e)] += ", '" + e + "'"
        else:
            result[len(e)] = "'" + e + "'"    

def show_values(result):
    """ Prints the strings with their respective lengths.

    Args:
        result (Dict): Dictionary with strings and their lengths.
    """
    for k, v in result.items():
        word = 's have' if ',' in v else ' has'
        print(''.join((f'The following string{word} {k} characters:\n', f'{v}')), '\n')
        
list_lenght(s_list)
show_values(result)  
```

```python
def initialize_matrix():
    """ Initialize an array with the length of the user's number.

    Returns:
        Matrix (String): Empty matrix.
    """
    return [[0 for _ in range(number)] for _ in range(number)]
```

```python
def build_matrix():
    """ Fill the matrix with the interleaved numbers (0 o 1).
    """
    counter = 0
    for row in range(number):
        for col in range(number):
            counter = 1 if counter == 0 else 0 
            matrix[row][col] = counter
```

```python
def show_matrix():
    """ Show the matrix.
    """
    for row in range(number):
        for col in range(number):
            print(str(matrix[row][col]).rjust(4), end='')
        print()
```

```python
items = {
        "Cafe": 7, 
        "Puma": 3, 
        "Vaso de papel": 2, 
        "Leche": 1, 
        "Coca-Cola": 4, 
        "Libro": 5,
    }

i = list(items.keys())
```

```python
def welcome_message():
    """ Show the welcome message.
    """
    print(f'Welcome, the store has the following supplies: {i[0]}', *i[1:], sep=', ')
    menu_message()
```

```python
def menu_message():
    """ Show the main menu.
    """
    print(''.join(('Please, select an operation:\n',
    '1. Check stock\n',
    '2. Storage\n',
    '3. Release\n',
    '4. Exit')))
```

```python
def get_number():
    """ Get a number from the user.

    Returns:
        int: user's number.
    """
    while True:
        welcome_message()
        print('\n')
        number = int(input('⚠ Please, enter a number: '))
        if 0 < number < 5: 
            break
        else:
            print('The number is not between 1 to 4. Please, try again.\n')
    return number
```

```python
def get_stock():
    """ Get an item from the user.

    Returns:
        String: user's item.
    """
    return input('⚠ Please, enter the item you want to check (item): ')
```

```python
def get_storage():
    """ Get an item and quantity from the user.

    Returns:
        String: user's item.
        int   : user's quantity.
    """
    return input('⚠ Please, enter the item and quantity you want to add (item quantity): ').split()
```

```python
def get_release():
    """ Get an item and quantity from the user.

    Returns:
        String: user's item.
        int   : user's quantity.
    """
    return input('⚠ Please, enter the item and quantity you want to remove (item quantity): ').split()
```

```python
def get_exit():
    """ Show the exit message.
    """
    return print('You have exited the program. ')
```

```python
def dict_operations(item, amount, operator):
    """ Perform an operation on the dictionary.

    Args:
        item (String) : User's item.
        amount (int)  : User's quantity.
        operator (int): Operation to do with the dictionary: 
                        show, add and/or delete.
    """
    exists = False
    for k, v in items.items():
        if k.lower() == item.lower() and operator == 1: 
            print(f'\n ➔ The stocks for {k} are {v} \n')
            exists = True
        if k.lower() == item.lower() and operator == 2:
            items[k] = v + int(amount)
            exists = True
        if k.lower() == item.lower() and operator == 3:
            exists = True
            if v - int(amount) <= 0:
                print(''.join(('\n ➔ The quantity desired is greater than the quantity stored.',
                    ' Please try another value for this transaction. \n')))
            else:
                items[k] = v - int(amount)
    if exists == False: 
        print('\n ➔ The item provided does not exist.\n')
```

```python
def set_operations():
    """ Call other functions.
    """
    while True:
        number = get_number()
        if number == 1:
            item = get_stock()
            dict_operations(item, None, 1)
        elif number == 2:
            item, amount = get_storage()
            dict_operations(item, amount, 2)
        elif number == 3:
            item, amount = get_release()
            dict_operations(item, amount, 3)
        else:
            break
        print('\n')
```

✅ Operations show, add and remove on the amounts of coffee.

```python
# Unity 10 - Q1

prime_list = []
minimun    = 2
maximun    = 10

def build_list():
    """ Build a list with prime numbers.

    Returns:
        List(int): prime numbers list.
    """
    for num in range(minimun, maximun + 1):
        if num > 1:
            for i in range(2, int(num / 2) + 1):
                if num % i == 0: break
            else:
                prime_list.append(num)
    return prime_list

def show_list():
    """ Show the list with prime numbers.
    """
    print('Prime list:')
    [print(num) for num in prime_list]
    
build_list()
show_list()
```

```python
# Unity 12 - Q2

sales  = (100, 121, 120, 130, 140, 120, 122, 123, 190)
before = sales[0]

def check_sales(before):
    """ Find the days that sales decreased.

    Args:
        before (int): The sale number of the previous day.
                        
    Returns:
        int: Number of days sales were down.
    """
    count = 0
    for i, s in enumerate(sales):
        if before > s:
            print(f'The value {s} has reduced in the day { i + 1 }.')
            count += 1
        before = s
    return count

count = check_sales(before)

print(f'In the last {len(sales)} days, sales have decreased by {count} days compared to the previous day.')
```

Escriba los resultados computacionales de los dos conjuntos siguientes. Encuentra los resultados de 1) a 7).

  <td>  s1 = {10, 20, 30, 40}<br>
        s2 = {30, 40, 50, 60, 70}<br>
        1) s1 | s2<br>
        2) s1 & s2<br>
        3) s1 – s2<br>
        4) s1 ^ s2<br>
        5) s1.issubset(s2)<br>
        6) s1.issuperset(s2)<br>
        7) s1.isdisjoint(s2)</td>

```python
# Unity 16 - Q2

"""
First Case  : s1 | s2
              Join the elements of two sets by removing the repeated elements in a new set
              [10, 20, 30, 40, 50, 60, 70]
             
Second Case : s1 & s2
              Join the common elements of two sets into a new set
              [30, 40]
             
Thrid Case  : s1 - s2
              Select the elements of s1 that s2 does not have in a new set
              [10, 20]
             
              s2 - s1
              [50, 60, 70]
              Select the elements of s2 that s1 does not have in a new set
             
Fourth Case : s1 ^ s2
              [10, 20, 50, 60, 70]
              Select the non-common elements of s1 and s2 into a new set

Fifth Case  : s1.issubset(s2)
              False
              There are no consecutive non-repeating elements in s1 and s2.

Sixth Case  : s1.issuperset(s2)
              False
              There are no consecutive non-repeating elements in s1 and s2.
             
Seventh Case: s1.isdisjoint(s2)
              False
              There are no consecutive non-repeating elements in s1 and s2.
"""

s1 = {10, 20, 30, 40}
s2 = {30, 40, 50, 60, 70}

print('s1 | s2: ', sorted(s1 | s2))
print('s1 & s2: ', sorted(s1 & s2))
print('s1 - s2: ', sorted(s1 - s2))
print('s2 - s1: ', sorted(s2 - s1))
print('s1 ^ s2: ', sorted(s1 ^ s2))

print('s1 is subset of s2: ', s1.issubset(s2))
print('s1 is uperset of s2: ', s1.issuperset(s2))
print('s1 is disjoint of s2: ', s1.isdisjoint(s2))

print('Other cases:')

# All elements of one set are in the other set.
s1 = {10, 20, 30, 40}
s2 = {10, 20, 30, 40, 70}

print('s1 is subset of s2: ', s1.issubset(s2)) # True
print('s2 is subset of s1: ', s2.issubset(s1)) # False

# All elements of one set are in the other set.
s1 = {10, 20, 30, 40}
s2 = {10, 20, 30, 40, 70}

print('s1 is uperset of s2: ', s1.issuperset(s2)) # False
print('s2 is uperset of s1: ', s2.issuperset(s1)) # True

# There are consecutive elements in the two sets.
s1 = {90, 20, 30, 40}
s2 = {50, 60, 70, 80, 10}

print('s1 is disjoint of s2: ', s1.isdisjoint(s2)) # True
print('s2 is disjoint of s1: ', s2.isdisjoint(s1)) # True
```
