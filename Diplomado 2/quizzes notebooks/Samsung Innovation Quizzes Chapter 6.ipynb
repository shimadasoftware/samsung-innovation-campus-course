<font color='blue'>Q1.</font>  Funci√≥n para proporcionar el m√≠nimo cambio de monedas



```python
def coin_change(coins, amount):
    """
    Get the least number of exchanges of coins of a value of money.

    Parameters
    ----------
    coins : List
        List of currencies available to provide a change.
    amount : int
        Amount of money to get a change.

    Returns
    -------
    List: Lists with exchange currencies.

    """
    changes = []
    largest = 0
    while (amount > 0):
        try:
            if amount - coins[largest] >= 0:
                changes.append(coins[largest])
                amount -= coins[largest]
                coins.pop(largest)
            else:
                largest += 1
        except IndexError:
            print(f'‚ö† There is NO currency to exchange: {amount}')
            break
    return changes
```

‚úÖ coins: [500, 100, 50, 50, 50, 10, 10]

‚úÖ amount: ‚Ç©710

üñ•Ô∏è Output: [500, 100, 50, 50, 10] 

```python
coins   = [500, 100, 50, 50, 50, 10, 10]
amount  = int(input('‚ö† Please, enter amount to change: '))
changes = coin_change(coins, amount)
print('Amount entered:', amount)
print(f'The number of coins are {len(changes)} which are {changes} ')
```

‚ùóEn clase el docente modific√≥ el quiz para realizarlo con monedas Colombianas. Por lo tanto, se manejan:
    


üîç **Without 700 pesos coin**

‚úÖ coins: [1000, 500, 200, 100, 50]

‚úÖ amount: 7750 pesos colombianos

üñ•Ô∏è Output: [1000, 1000, 1000, 1000, 1000, 1000, 1000, 500, 200, 50] 

```python
def coin_change(coins, amount):
    """
    Get the least number of exchanges of coins of a value of money.

    Parameters
    ----------
    coins : List
        List of currencies available to provide a change.
    amount : int
        Amount of money to get a change.

    Returns
    -------
    List: Lists with exchange currencies.

    """
    changes = []
    largest = 0
    while (amount > 0):
        try:
            if amount < coins[largest]:
                largest += 1
            else:
                changes.append(coins[largest])
                amount -= coins[largest]
        except IndexError:
            print(f'‚ö† There is NO currency to exchange: {amount}')
            break
    return changes
```

```python
coins   = [1000, 500, 200, 100, 50]
amount  = int(input('‚ö† Please, enter amount to change: '))
changes = coin_change(coins, amount)
print('Amount entered:', amount)
print(f'The number of coins are {len(changes)} which are {changes} ')
```
