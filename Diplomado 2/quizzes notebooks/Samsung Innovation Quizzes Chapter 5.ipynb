```python
def find_anagram(w1, w2):
    """
    Find an anagram in two words.

    Parameters
    ----------
    w1 : String
        First word.
    w2 : String
        Second word.

    Returns
    -------
    None.

    """
    anagram, word2 = '', w2
    if len(w1) == len(w2): 
        for letter in w1:
            if letter in word2:
                anagram += letter
                ind   = word2.index(letter)
                word2 = word2[ : ind] + word2[ind + 1 : ]
        print(f"{w1.capitalize()} and {w2.capitalize()} are anagrams." if w1 == anagram else f"{w1.capitalize()} and {w2.capitalize()} are NOT anagrams.")
    else:
        print(f"{w1} and {w2} are NOT anagrams.")
```

```python
def selection_sort2(w1, w2):
    """
    Find an anagram in two list.

    Parameters
    ----------
    w1 : List
        First word.
    w2 : List
        Second word.

    Returns
    -------
    boolean
        True: If it is an anagram.
        False: If it is NOT an anagram.

    """
    n = len(w1)
    if len(w1) == len(w2):
        for i in range(n - 1):
            print(f'Second word {w2}')
            for j in range(n):
                if w1[i] == w2[j]:
                    w2[j], w2[i] = w2[i], w2[j]
        print(f'Second word {w2}')
        return True if w1 == w2 else False
    return False
```

```python
def sort_by_insert2(word):
    """
    Sort a list according to the ASCII values of each word.

    Parameters
    ----------
    word : List
        Word to order.

    Returns
    -------
    List:
        The word to ordained.

    """
    size = len(word)
    for i in range(1, size):
        x = word[i]
        j = i - 1
        while j >= 0 and word[j] > x:
            word[j + 1] = word[j]
            j -= 1
        word[j + 1] = x
    return word
```

```python
def is_anagram(w1, w2):
    """
    Find an anagram in two list.

    Parameters
    ----------
    w1 : List
        First word.
    w2 : List
        Second word.

    Returns
    -------
    None.

    """
    cp1, cp2 = w1.copy(), w2.copy()
    w1 = sort_by_insert2(w1)
    w2 = sort_by_insert2(w2)
    
    if w1 == w2:
        print(f"{cp1} and {cp2} are anagrams.")
    else:
        print(f"{cp1} and {cp2} are NOT anagrams.")
```

üîç **Bubble algorithm**

```python
def bubble(num):
    """
    Sort numbers in ascending order in a list.

    Parameters
    ----------
    num : List
        List of numbers.

    Returns
    -------
    List: Ordered numbers list.

    """
    count, size = 0, len(num)
    for i in range(size):
        print(num)
        for j in range(size - 1):
            if num[j] > num[j + 1]:
                count += 1
                num[j], num[j + 1] = num[j + 1], num[j]
    return num
```

```python
def multiple_fusion(numbers_list):
    """
    Unify list of lists into one list.

    Parameters
    ----------
    numbers_list : List
        List of numbers.

    Returns
    -------
    List: Ordered numbers list.

    """
    numbers = []
    numbers = [y for x in numbers_list for y in x]
    return bubble(numbers)
```

üîç **Selection Sort algorithm**

```python
def selection_sort(num):
    """
    Sort numbers in ascending order in a list.

    Parameters
    ----------
    num : List
        List of numbers.

    Returns
    -------
    List: Ordered numbers list.

    """
    n = len(num)
    for i in range(n - 1):
        print(num)
        smallest = i
        for j in range(i + 1, n):
            if num[j] < num[smallest]:
                smallest = j
        num[i], num[smallest] = num[smallest], num[i]
    return num
```

```python
def multiple_fusion(numbers_list):
    """
    Unify list of lists into one list.

    Parameters
    ----------
    numbers_list : List
        List of numbers.

    Returns
    -------
    List: Ordered numbers list.

    """
    numbers = []
    numbers = [y for x in numbers_list for y in x]
    return selection_sort(numbers)
```

üîç **First algorithm**

```python
def partition(nums, low, high):
    """
    Split elements from smallest to largest based on a pivot.

    Parameters
    ----------
    nums : List
        List to order.
    low : int
        Low range.
    high : int
        High range.

    Returns
    -------
    int: Pivot end position.

    """
    pivot = nums[low]
    left, right = low + 1, high
    while left < right:
        while left <= right and nums[left] <= pivot:
            left += 1
        while left <= right and nums[right] >= pivot:
            right -= 1
        if left < right:
            nums[left], nums[right] = nums[right], nums[left]
    pivot_point = right
    nums[low], nums[pivot_point] = nums[pivot_point], nums[low]
    return pivot_point
```

```python
def quick_sort(nums, low, high):
    """
    Sorts the pivot within the list, having as pivot a random element of the list.

    Parameters
    ----------
    nums : List
        List to order.
    low : int
        Low range.
    high : int
        High range.

    Returns
    -------
    None.

    """
    if low < high:
        pivot_point = partition(nums, low, high)
        quick_sort(nums, low, pivot_point - 1)
        quick_sort(nums, pivot_point + 1, high)
    return nums[-1]
```

üîç **Second algorithm**

```python
def partition_2(nums, low, high):
    """
    Split elements from smallest to largest based on a pivot.

    Parameters
    ----------
    nums : List
        List to order.
    low : int
        Low range.
    high : int
        High range.

    Returns
    -------
    int: Pivot end position.

    """
    rand = randint(low, high)
    nums[low], nums[rand] = nums[rand], nums[low]
    pivot, left, right = nums[low], low, high
    while left < right:
        while left < high and nums[left] <= pivot:
            left += 1
        while right > low and pivot <= nums[right]:
            right -= 1
        if left < right:
            nums[left], nums[right] = nums[right], nums[left]
    nums[low], nums[right] = nums[right], nums[low]
    return right
```

```python
def quick_sort_2(nums, low, high):
    """
    Sorts the pivot within the list, having as pivot a random element of the list.

    Parameters
    ----------
    nums : List
        List to order.
    low : int
        Low range.
    high : int
        High range.

    Returns
    -------
    None.

    """
    if low < high:
        pivot_point = partition_2(nums, low, high)
        quick_sort_2(nums, low, pivot_point - 1)
        quick_sort_2(nums, pivot_point + 1, high)
    return nums[-1]
```

```python
def merge_sort_2(S, low, high):
    """
    Sort a list of numbers by ranges

    Parameters
    ----------
    S : List
        List to order.
    low : int
        Low range.
    high : int
        High range.

    Returns
    -------
    None.

    """
    global count 
    count += 1
    if low < high:
        mid = (low + high) // 2
        merge_sort_2(S, low, mid)
        merge_sort_2(S, mid + 1, high)
        fusion_2(S, low, mid, high)
```

```python
def fusion_2(S, low, mid, high):
    """
    Sort the list from left to right.

    Parameters
    ----------
    S : List
        List to order.
    low : int
        Low range.
    mid : int
        Middle value.
    high : int
        High range.

    Returns
    -------
    None.

    """
    R = []
    i, j = low, mid + 1
    while i <= mid and j <= high:
        if S[i] < S[j]:
            R.append(S[i])
            i += 1
        else:
            R.append(S[j])
            j += 1
    if i > mid:
        for k in range(j, high + 1):
            R.append(S[j])
    else:
        for k in range(i, mid + 1):
            R.append(S[i])
    for k in range(len(R)):
        S[low + k] = R[k]
```
